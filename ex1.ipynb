```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ESERCIZIO 1
A = np.array([
    [0,0,1,1/2],
    [1/3,0,0,0],
    [1/3,1/2,0,1/2],
    [1/3,1/2,0,0]
])

A_eigenvalues, A_eigenvectors = np.linalg.eig(A)
fig, ax = plt.subplots(1,2, figsize=(14,4))

# ------------------------------------------------------
# AUTOVALORI: reali, ordinati in ordine decrescente
# ------------------------------------------------------
real_eigs = A_eigenvalues.real
sorted_eig_idx = np.argsort(real_eigs)[::-1]
sorted_eigs = real_eigs[sorted_eig_idx]

ax[0].bar(range(len(sorted_eigs)), sorted_eigs)
ax[0].set_title("Autovalori reali (ordinati decrescenti)")
ax[0].set_xlabel("Indice autovalore ordinato")
ax[0].set_ylabel("Valore")
ax[0].set_xticks(range(len(sorted_eigs)))

# ------------------------------------------------------
# AUTOVETTORE DOMINANTE: ordinato decrescente
# ------------------------------------------------------
# indice dellâ€™autovalore dominante
dominant_index = sorted_eig_idx[0]
dominant_vec = A_eigenvectors[:, dominant_index].real
dominant_vec = dominant_vec/np.sum(dominant_vec)


# ordina il PageRank in ordine decrescente
pr_sorted_idx = np.argsort(dominant_vec)[::-1]
pr_sorted_values = dominant_vec[pr_sorted_idx]

# bar plot con indice della pagina sotto
ax[1].bar(range(len(pr_sorted_values)), pr_sorted_values)
ax[1].set_title("PageRank (componenti autovettore dominante ordinate)")
ax[1].set_xlabel("Indice pagina")
ax[1].set_ylabel("Rank")

# Imposta i numeri delle pagine (etichette)
ax[1].set_xticks(range(len(pr_sorted_values)))
ax[1].set_xticklabels(pr_sorted_idx+1)

plt.tight_layout()
plt.show()
```

```python
B = np.array([
    [0,0,1/2,1/2,0],
    [1/3,0,0,0,0],
    [1/3,1/2,0,1/2,1],
    [1/3,1/2,0,0,0],
    [0,0,1/2,0,0]
])
B_eigenvalues, B_eigenvectors = np.linalg.eig(B)

fig, ax = plt.subplots(1,2, figsize=(14,4))

# ------------------------------------------------------
# AUTOVALORI: reali, ordinati in ordine decrescente
# ------------------------------------------------------
real_eigs = B_eigenvalues.real
sorted_eig_idx = np.argsort(real_eigs)[::-1]
sorted_eigs = real_eigs[sorted_eig_idx]

ax[0].bar(range(len(sorted_eigs)), sorted_eigs)
ax[0].set_title("Autovalori reali (ordinati decrescenti)")
ax[0].set_xlabel("Indice autovalore ordinato")
ax[0].set_ylabel("Valore")
ax[0].set_xticks(range(len(sorted_eigs)))

# ------------------------------------------------------
# AUTOVETTORE DOMINANTE (NORMALIZZATO): ordinato decrescente
# ------------------------------------------------------
# indice dellâ€™autovalore dominante
dominant_index = sorted_eig_idx[0]
dominant_vec = B_eigenvectors[:, dominant_index].real
dominant_vec = dominant_vec/np.sum(dominant_vec)

# ordina il PageRank in ordine decrescente
pr_sorted_idx = np.argsort(dominant_vec)[::-1]
pr_sorted_values = dominant_vec[pr_sorted_idx]
pr_sorted_values = pr_sorted_values

# bar plot con indice della pagina sotto
ax[1].bar(range(len(pr_sorted_values)), pr_sorted_values)
ax[1].set_title("PageRank (componenti autovettore dominante ordinate)")
ax[1].set_xlabel("Indice pagina")
ax[1].set_ylabel("Rank")

# Imposta i numeri delle pagine (etichette)
ax[1].set_xticks(range(len(pr_sorted_values)))
ax[1].set_xticklabels(pr_sorted_idx+1)

plt.tight_layout()
plt.show()

B_eigenvectors
```
